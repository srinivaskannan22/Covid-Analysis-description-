```python
import dash
from dash import html,dcc
from dash.dependencies import Output,Input
import dash_bootstrap_components as dbc
from dash_bootstrap_templates import load_figure_template 
from jupyter_dash import JupyterDash
app=JupyterDash(__name__,external_stylesheets=[dbc.themes.QUARTZ])
load_figure_template('quartz')
app.layout=dcc.Tabs([dcc.Tab(label='Case',children=[html.Div([dbc.Row(dbc.Col(html.H1('Covid Cases'),width='auto'),justify='center'

),dbc.Row([dbc.Col(html.Div([dcc.Dropdown(
    options=('Confirmed','Deaths'),id='cou',
    
),
html.H5('highest Content',style={'textAlign': 'center'}),
html.Div([
html.H5('Covid-Case',style={'color':'cyan','text-decoration':'underline'}),html.H6(id='hover',style={'display': 'flex',
         'justify-content':'flex-end','margin-top':'20px','font-family': 'Timenewsroman','font-weight':'bold'})
],style={'color':'orange',
         'width':'150px',
         'height':'150px',
         'textAlign': 'center',
         'margin': '0 auto',
         'border': '5px solid cyan',
         'background-color':'white',
         })]),width=6),
dbc.Col(dcc.Graph(id='cou2'),width=6)]),

]),

]),
dcc.Tab(label='case2',children=[
    html.Div([html.H2('Map',style={'textAlign':'center'}),
              dbc.Row([dbc.Col(dcc.RadioItems(
                  options=('usa','world'),id='mapselect'
              ),width=1),
              dbc.Col(
              
              dcc.Graph(id='map'),width=11)
              ])
              
              
              
              
              ]),



]),
dcc.Tab(label='case3',children=[
    html.Div([
        html.H2('Death/Recovered',style={'textAlign':'center'}),
        dcc.Dropdown(options=('Death','Recovered'),id='Death'),
        dbc.Row([dbc.Col(dcc.Graph(id='mapdeath',hoverData={'points':[{'customdata':['Date']}]})),
        dbc.Col(dcc.Graph(id='Mappie'))])


])

])


])
@app.callback(Output('cou2','figure'),
               Input('cou','value'))
def change(x2):
    if x2 is None:
        return {}
    fig=px.bar(fullgrouped.groupby('Country/Region').sum().reset_index().sort_values(by=x2,ascending=False).iloc[:10,:],x='Country/Region',y=x2,color='Country/Region',title=f'highest {x2}')
    return fig
@ app.callback(Output('hover','children'),
               Input('cou2','hoverData'))
def hov(y):
    if y is None:
        return None
    country = y['points'][0]['x']  
    value = y['points'][0]['y']        
    return f"Country: {country}, Value: {value}"
@app.callback(Output('map','figure'),
              Input("mapselect",'value')

)
def map1(x):
    if x is None:
        return {}
    elif x=='world':
        fig = px.scatter_mapbox(
            clead.groupby('Country/Region').agg({'Confirmed':'sum','Deaths':'sum','Lat':'mean','Long':'mean'}).sort_values(by='Deaths',ascending=False).reset_index().iloc[:100,:], 
            lat="Lat", 
            lon="Long", 
            color="Country/Region", 
            size="Deaths",
            color_continuous_scale=px.colors.cyclical.IceFire,
            zoom=0,
            mapbox_style="carto-positron"  
        )
        return fig
    else:
        fig = px.scatter_mapbox(
            usa_county.groupby('Province_State').agg({'Lat':'mean','Long_':'mean','Confirmed':'sum','Deaths':'sum'}).sort_values(by='Deaths',ascending=False).iloc[:50,:].reset_index(), 
            lat="Lat", 
            lon="Long_", 
            color="Province_State", 
            size="Deaths",
            color_continuous_scale=px.colors.cyclical.IceFire,
            zoom=2,
            mapbox_style="carto-positron"  
        )
        return fig
@app.callback([Output('mapdeath','figure'),Output('Mappie','figure')],
              Input('Death','value')

    )
def map3(x):
    if x is None:
        return {},{}
    if x:
        if x=="Death":
            fig=px.line(daywise.sort_values(by='Deaths / 100 Cases',ascending=False).iloc[:50,:].sort_values(by='Date',ascending=True),x='Date',y='Deaths / 100 Cases')
            fig2=px.pie(daywise.sort_values(by='Deaths / 100 Cases',ascending=False).iloc[:10,:].sort_values(by='Date',ascending=True),names='Date',values='Deaths / 100 Cases',hole=0.7)
        
            
        else :
            fig=px.bar(daywise.sort_values(by='Recovered / 100 Cases',ascending=False).iloc[:200,:].sort_values(by='Date',ascending=True),x='Date',y='Recovered / 100 Cases')
            fig2=px.pie(daywise.sort_values(by='Recovered / 100 Cases',ascending=False).iloc[:10,:].sort_values(by='Date',ascending=True),names='Date',values='Recovered / 100 Cases',hole=0.7)
        
    return fig,fig2        


    
    
    
      



        

    
if __name__=="__main__":
    app.run_server(debug=True,port=8590)
```
